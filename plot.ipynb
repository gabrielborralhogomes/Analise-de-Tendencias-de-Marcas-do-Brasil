```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
```

# Carregar dados

```python
df_Regiao = pd.read_csv("dados/Marcas_Regiao.csv")
df_Regiao.set_index('Marca', inplace=True)
df_Regiao = df_Regiao.T  # Transpor para ter as datas no eixo x

df_Tempo = pd.read_csv("dados/Marcas_Tempo.csv")
df_Tempo.set_index('Marca', inplace=True)
df_Tempo = df_Tempo.T  # Transpor para ter as datas no eixo x
```

## Somando as linhas

```python
Total_Tempo = {}
for index, row in df_Tempo.iterrows():
    # Somando os valores da linha, excluindo a coluna 'Marca'
    total_sum = row.sum()  # Isso assume que todas as colunas a partir do índice 1 são numéricas

    # Preenchendo o dicionário com o índice e o resultado da soma
    Total_Tempo[index] = total_sum  # Usando index como chave e total_sum como valor

# Exibindo o dicionário com os totais
print(Total_Tempo)
```

```python
Total_Regiao = {}
for index, row in df_Regiao.iterrows():
    # Somando os valores da linha, excluindo a coluna 'Marca'
    total_sum = row.sum()  # Isso assume que todas as colunas a partir do índice 1 são numéricas

    # Preenchendo o dicionário com o índice e o resultado da soma
    Total_Regiao[index] = total_sum  # Usando index como chave e total_sum como valor

# Exibindo o dicionário com os totais
print(Total_Regiao)
```

## Somando as colunas

```python
# Somando os valores de cada coluna, excluindo a coluna 'Marca'
Total_Marca_Tempo = {}
for column_name, column_data in df_Tempo.items():
    total_sum = column_data.sum()  # Somando a coluna
    Total_Marca_Tempo[column_name] = total_sum  # Adicionando ao dicionário

print(Total_Marca_Tempo)
```

```python
# Somando os valores de cada coluna, excluindo a coluna 'Marca'
Total_Marca_Reigao = {}
for column_name, column_data in df_Regiao.items():
    total_sum = column_data.sum()  # Somando a coluna
    Total_Marca_Reigao[column_name] = total_sum  # Adicionando ao dicionário

print(Total_Marca_Reigao)
```

# Plotagem

```python
# Supondo que df_Regiao já esteja carregado e tenha as regiões como índice
# Criando um dicionário para armazenar a marca com o maior valor por região
max_brands = {}

# Iterando sobre cada linha (região) para encontrar a marca com o maior valor
for index, row in df_Regiao.iterrows():
    # Encontrando a marca com o maior valor na linha
    max_value = row.max()
    max_brand = row.idxmax()  # Marca correspondente ao maior valor
    max_brands[index] = max_brand

# Criando um DataFrame a partir do dicionário
df_max_brands = pd.DataFrame(list(max_brands.items()), columns=['Região', 'Marca'])

# Carregando o mapa do Brasil
# Altere o caminho para o local onde você descompactou os dados
shapefile_path = "caminho/para/seu/shapefile/ne_110m_admin_1_states_provinces.shp"
brasil = gpd.read_file(shapefile_path)

# Filtrando apenas os estados do Brasil
brasil = brasil[brasil['admin'] == "Brazil"]

# Unindo os dados da marca com o mapa do Brasil
# (Você pode precisar ajustar os nomes das regiões para corresponder)
brasil = brasil.merge(df_max_brands, left_on='name', right_on='Região', how='left')

# Plotando o mapa
fig, ax = plt.subplots(1, 1, figsize=(12, 12))
brasil.boundary.plot(ax=ax, linewidth=1)
brasil.plot(column='Marca', ax=ax, legend=True,
            legend_kwds={'bbox_to_anchor': (1, 1), 'loc': 'upper left'})

plt.title('Marca com Maior Valor por Região do Brasil')
plt.show()
```
